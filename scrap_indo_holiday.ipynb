```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
holidays=[]

for i in range (15,25): #2015-2024
    url = f'https://www.timeanddate.com/holidays/indonesia/20{i}'
    page = requests.get(url)
    page = page.content
    soup = BeautifulSoup(page, 'html.parser')

    rows = soup.find_all('tr')
    for row in rows:
        
        if row.get('data-date') is not None:
            public_date =row.find('th',class_ = 'nw').text
            public_day =row.find('td', class_='nw').text
            public_name =row.find('a').text
            public_type =row.find_all('td')[-1].text

        year_sentence = soup.find ('h2', id = 'holidays').text
        year = year_sentence[-4:]
        
        print(year)
        print(public_date)
        print(public_day)
        print(public_name)
        print(public_type)
        print ('')

        holidays.append((year,public_date, public_day, public_name,public_type))

    print ('-------------------------------')
    
            
    
```

```python
df = pd.DataFrame(holidays, columns=['Year', 'Date', 'Day', 'Name','Type'])
df.to_csv('Indonesia Public Holiday 2015 - 2024.csv')
```
